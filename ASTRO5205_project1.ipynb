# Recreating the plots with solar system planetary data

The data has been taked from NASA's [Planetary Fact Sheet](https://nssdc.gsfc.nasa.gov/planetary/factsheet/).

```python
# Creating arrays to store NASA Planetary Data arranged in increasing distance from the Sun
# [Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune]

# Masses of planets
solsys_mass = [0.330e24, 4.87e24, 5.97e24, 0.642e24,
               1898e24, 568e24, 86.8e24, 102e24]*u.kg
solsys_masse = solsys_mass.to(u.earthMass)

# Periods of planets
solsys_orbper = [88.0, 224.7, 365.2, 687, 4331, 10747,
                 30589, 59800]*u.day

# Semi-major Axes of planets
solsys_orbdist = [57.9e6, 108.2e6, 149.6e6, 228.0e6, 778.5e6,
                  1432.0e6, 2867.0e6, 4515.0e6]*u.km
solsys_orbsmax = solsys_orbdist.to(u.au)

# Diameters and Radii of the planets
solsys_diameter = [4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528]
solsys_rad = [i/2 for i in solsys_diameter]*u.km
solsys_rade = solsys_rad.to(u.earthRad)
```

## I. 3. Detection Method Sensitivties

```python
# Transit Sensititivy
# (R_p / R_Earth)^2 = 3 * (P / T)^(1/2) ==> R_p = R_Earth * { 3 * (P / T)^(1/2) }^(1/2)
T = u.year
R_Earth = u.earthRad
P = np.arange(1,1e4,10) * u.day
R_p = R_Earth * np.sqrt(3 * np.sqrt(P / T))
```

```python
# Mass vs. Period

# Creating plot
fig, axs = plt.subplots(1, 1, figsize=(8, 6))

# Plotting data
axs.scatter(data['pl_orbper'], data['pl_bmasse'], marker='.', color='black', label='exoplanets')
axs.scatter(solsys_orbper, solsys_masse, marker='.', color='red', label='solar system planets')
axs.plot(P, R_p, color="blue", label="Transit sensitivity")

# Setting labels
axs.set_xlabel('Period [days]')
axs.set_ylabel('Mass [M$_{Earth}$]')
axs.set_title('Mass vs. Period')

# Adding scales to axes
axs.set_xscale('log')
axs.set_yscale('log')

# Adding axis limits
axs.set_xlim(1e-1, 1e5)

# Adding a legend
axs.legend()

plt.tight_layout()  # Adjust layout for better spacing
axs.grid('on') # Show grid
plt.show()

#p_arr = 10**np.arange(0,3,0.01) * u.day # An array of orbital period ranging from 1 to 1000 days


#i = np.where((data["default_flag"] == 1) & (data["discoverymethod"] == "Transit"))

#plt.plot(data["pl_orbper"][i], data["pl_radj"][i], "b.", label="Transiting Planets")
```
